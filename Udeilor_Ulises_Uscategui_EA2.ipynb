## Realiza una limpieza de datos de las variables anteriores, haciendo uso de .zip(), .map(), .replace(), .fillna(), etc. (10 puntos).

Con replace vamos a remplazar los valores repetidos en Estado

```python
# remplazando datos con replace

df_covid['Estado'] = df_covid['Estado'].replace(["LEVE", "leve"], "Leve")
df_covid["Estado"].value_counts(dropna=False)
```

Con replace vamos a remplazar los valores repetidos en Sexo

```python
df_covid['Sexo'] = df_covid['Sexo'].replace(["f"], "F")
df_covid['Sexo'] = df_covid['Sexo'].replace(["m"], "M")
df_covid["Sexo"].value_counts(dropna=False)
```

Ahora a remplazazar los valores de nombre del pais

```python
print(df_covid["Nombre del país"].unique())
```

```python
df_covid['Nombre del país'] = df_covid['Nombre del país'].replace(["ISLAS VÍRGENES DE LOS ESTADOS UNIDOS",'PUERTO RICO', "ESTADOS UNIDOS"], "ESTADOS UNIDOS DE AMÉRICA")
df_covid['Nombre del país'] = df_covid['Nombre del país'].replace(['REPÚBLICA DOCIMINCANA'], "REPÚBLICA DOMINICANA")
df_covid['Nombre del país'] = df_covid['Nombre del país'].replace(['PANAMA'], 'PANAMÁ')
df_covid['Nombre del país'] = df_covid['Nombre del país'].replace(['PERU'], 'PERÚ')
df_covid['Nombre del país'] = df_covid['Nombre del país'].replace(['VENEUELA'], 'VENEZUELA')
df_covid['Nombre del país'] = df_covid['Nombre del país'].replace(['MEXICO'], 'MÉXICO')
df_covid['Nombre del país'] = df_covid['Nombre del país'].replace(['SANTA HELENA','REINO UNIDO'], 'REINO UNIDO DE GRAN BRETAÑA E IRLANDA DEL NORTE')
df_covid['Nombre del país'] = df_covid['Nombre del país'].replace(['GUADALUPE','POLINESIA FRANCESA', 'SAN BARTOLOMÉ' ], 'FRANCIA')
print(df_covid["Nombre del país"].unique())
```

```python
Pais_limpiar=df_covid["Nombre del país"].unique()
```

```python
Pais_limpiar = np.array(Pais_limpiar)
#eliminar registros nulos
Pais_limpiar = Pais_limpiar[~pd.isnull(Pais_limpiar)]

# Convertir a una lista ordenada alfabeticamente
Pais_limpiar = np.array(sorted(Pais_limpiar))

Pais_limpiar
```

```python
df_covid['Nombre del país'] = df_covid['Nombre del país'].replace(['TABATINGA'], 'BRASIL')
print(df_covid["Nombre del país"].unique())
```

# Identifica registros duplicados. (5 puntos).

```python
# identificar registros dulicados basados en todas las columnas
duplicates = df_covid_copy[df_covid_copy.duplicated(keep=False)]

# Mostrar en pantalla los registros duplicados.
duplicates
```

No existen registros duplicados en la base de datos

```python
# identificar valores atipicos
numeric_cols = df_covid_copy.select_dtypes(include=np.number).columns

# Iterar sobre las columnas numéricas
for col in numeric_cols:
    # Calcular el rango intercuartílico (IQR)
    Q1 = df_covid_copy['Edad'].quantile(0.25)
    Q3 = df_covid_copy['Edad'].quantile(0.75)
    IQR = Q3 - Q1

    # Definir los límites para identificar valores atípicos
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Identificar valores atípicos
    outliers = df_covid_copy[(df_covid_copy['Edad'] < lower_bound) | (df_covid_copy[col] > upper_bound)]

    # Imprimir los valores atípicos para la columna actual
    print(f"Valores atípicos en la columna '{'Edad'}':")
    print(outliers[['Edad']])
    print("-" * 20)
```

```python

# Udentificando valores atipicos de variables categoricas,
categorical_cols = df_covid_copy.select_dtypes(exclude=np.number).columns
for col in categorical_cols:
    value_counts = df_covid_copy['Nombre departamento'].value_counts(normalize=True)
    rare_values = value_counts[value_counts < 0.01] # Ejemplo: valores que aparecen en menos del 1% de los datos
print(f"Valores poco frecuentes en '{'Nombre departamento'}':")
print(rare_values)
print("-"*20)
```

Los valores mas atipicos son Putumayo, Tolima, caldas y Cundinamarca

```python
# Udentificando valores atipicos de variables categoricas,
categorical_cols = df_covid_copy.select_dtypes(exclude=np.number).columns
for col in categorical_cols:
    value_counts = df_covid_copy['Nombre municipio'].value_counts(normalize=True)
    rare_values = value_counts[value_counts < 0.01] # Ejemplo: valores que aparecen en menos del 1% de los datos
print(f"Valores poco frecuentes en '{'Nombre municipio'}':")
print(rare_values)
print("-"*20)
```

Los valores mas atipicos son Gameza, Guepsa, Pensilvania, Puerto Colombia, Barrancabermeja

```python
# Udentificando valores atipicos de variables categoricas (pais),
categorical_cols = df_covid_copy.select_dtypes(exclude=np.number).columns
for col in categorical_cols:
    value_counts = df_covid_copy['Nombre del país'].value_counts(normalize=True)
    rare_values_country = value_counts[value_counts < 0.01] # Ejemplo: valores que aparecen en menos del 1% de los datos
print(f"Valores poco frecuentes en '{'Nombre del país'}':")
print(rare_values_country)
print("-"*20)
```
