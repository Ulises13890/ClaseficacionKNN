Importamos las librerias correspondientes

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.metrics import confusion_matrix
```

Seleccionamos los datos que van a ir en X, y de la funcion

```python
X=dfusers.iloc[ : , 2:4].values
y=dfusers.iloc[: ,4]. values
print(X)
print(y)
```

Seleccion de datos de entrenamiento y de prueba

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test=train_test_split(X,y,test_size=0.25, random_state=9)
```

```python
#do the training with sklearn
from sklearn.preprocessing import StandardScaler
sc_X=StandardScaler()
#se usa fit_transform para escalar los datos de la serie X_train y X_test
X_train=sc_X.fit_transform(X_train)
X_test=sc_X.fit_transform(X_test)
```

```python
print(X_train)
print(X_test)
#Verificamos el escalado de los conjuntos de datos con valores con una media de 0 y desviacion estandar de 1 para que los 
#pueda procesar el algoritmo de apremdizaje automatico
```

```python
from sklearn.neighbors import KNeighborsClassifier
Classifier=KNeighborsClassifier(n_neighbors=5, metric="minkowski",p=2)
Classifier.fit(X_train,y_train)
```

```python
#ponemos el modelo a predecir los valores de X_test
y_pred=Classifier.predict(X_test)
#Creamos una matriz de confusion para probar el error de la matriz de prediccion vs la matriz de prueba
cm=confusion_matrix(y_pred,y_test)
print(cm)
```

Esta matriz de confusion dice que 58 valores verdaderos se predicen correctamente verdaderos, 
5 valores falsos se predicen incorrectamente como verdaderos
2 valores verdaderos se predicen incorrectamente como falsos 
35 valores falsos se predicen correctamente como falsos

Si se suman los resultados correctos con los resultados incorrectos nos da una precision de 99.92%
